```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'src.png'
file_dst = 'dst_png'

img_src = cv2.imread(file_src,1)#入力画像の読み込み

#グレースケール変換
img_gray = cv2.cvtColor(img_src,cv2.COLOR_BGR2GRAY)

#ヒストグラム表示用,100x256ピクセル,0で初期化
img_hst = np.zeros([100,256]).astype('uint8')
rows, cols = img_hst.shape[:2]
img_dst = cv2.equalizeHist(img_gray)

#度数分布を求める
hdims=[256]
hranges=[0,256]
hist=cv2.calcHist([img_dst],[0],None,hdims,hranges)
#hist=cv2.calcHist([img_gray],[0],None,hdims,hranges)

#度数の最大値を取得
min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(hist)
max_val=1100
for i in range(0,256):
    v=hist[i]
    print(v)
    cv2.line(img_hst,(i,rows),(i,rows-rows*(v/max_val)),(255,255,255))

#ネガポジ変換
img_dst = 255-img_src

#表示
cv2.imshow("src Show Input Image", img_src)#入力画像表示
cv2.imshow("dst Show GRAY Image", img_dst)#出力画像表示
cv2.imshow("hst Show GRAY Image", img_hst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'jingu_s.png'
file_dst = 'dst_png'

img_src = cv2.imread(file_src,1)#入力画像の読み込み

#ルックアップテーブルと関数LUT
shift = 100
table = np.arange(256, dtype = np.uint8)
for i in range(0,255):
    j = i + shift
    if j < 0:
        table[i] = 0
    elif j > 255:
        table[i] = 255
    else:
        table[i] = j
img_dst = cv2.LUT(img_src,table)

#表示
cv2.imshow("src Show Input Image", img_src)#入力画像表示
cv2.imshow("dst Show GRAY Image", img_dst)#出力画像表示

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'jingu_s.png'
file_dst = 'dst_png'

img_src = cv2.imread(file_src,1)#入力画像の読み込み

#ルックアップテーブルと関数LUT
#コントラスト強調
min = 50
max = 250
table = np.arange(256, dtype = np.uint8)
for i in range(0, min):
    table[i] = 0
for i in range(min,max):
    table[i] = 255 * (i-min )/ (max-min)
for i in range(max,255):
    table[i] = 255

img_dst = cv2.LUT(img_src,table)

#表示
cv2.imshow("src Show Input Image", img_src)#入力画像表示
cv2.imshow("dst Show GRAY Image", img_dst)#出力画像表示

cv2.waitKey(0)
cv2.destroyAllWindows()
```
