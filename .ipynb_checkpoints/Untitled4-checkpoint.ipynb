```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'jingu_s.png'
file_dst = 'dst_png'

img_src = cv2.imread(file_src,1)#入力画像の読み込み

#ルックアップテーブルと関数LUT
#コントラスト強調
min = 190
max = 200
table = np.arange(256, dtype = np.uint8)
for i in range(0, min):
    table[i] = 0
for i in range(min,max):
    table[i] = 255 * (i-min )/ (max-min)
for i in range(max,255):
    table[i] = 255
img_dst = cv2.LUT(img_src,table)
#表示
cv2.imshow("src Show Input Image", img_src)#入力画像表示
cv2.imshow("dst Show color Image", img_dst)#出力画像表示

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'morph.png'
file_dst = 'dst.png'

img_src = cv2.imread(file_src,0)#入力画像の読み込み

element8 = np.array([[1,1,1],[1,1,1],[1,1,1]],np.uint8) #8近傍
#オープニング
img_tmp = cv2.morphologyEx(img_src,cv2.MORPH_OPEN, element8)
#クロージング
img_dst = cv2.morphologyEx(img_tmp,cv2.MORPH_CLOSE, element8)

#表示
cv2.imshow("src Show Input Image", img_src)#入力画像表示
cv2.imshow("dst Show morphorogy Image", img_dst)#出力画像表示

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'daruma.png'
file_dst = 'dst.png'

img_src = cv2.imread(file_src,1)#入力画像の読み込み
img_gray =cv2.cvtColor(img_src,cv2.COLOR_BGR2GRAY)

thresh = 100
ret, img_dst = cv2.threshold(img_gray,thresh,255,cv2.THRESH_BINARY)

cv2.imshow("dst Show 二値化 Image",img_dst)

contours =cv2.findContours(img_dst,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[1]
area =  cv2.contourArea(contours[0])
perimeter = cv2.arcLength(np.array(contours[0]),True)
roundness = 4*np.pi*area/perimeter/perimeter


#表示

print("面積",area)
print("周囲長",perimeter)
print("円形度"roundness)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'daruma.png'

img_src = cv2.imread(file_src,1)#入力画像の読み込み
img_gray =cv2.cvtColor(img_src,cv2.COLOR_BGR2GRAY)

thresh = 100
ret, img_dst = cv2.threshold(img_gray,thresh,255,cv2.THRESH_BINARY)

cv2.imshow("dst Show 二値化 Image",img_dst)

contours=cv2.findContours(img_dst,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[0]
area=cv2.contourArea(contours[0])
perimeter = cv2.arcLength(np.array(contours[0]),True)
roundness = 4*np.pi*area/perimeter/perimeter


#表示
print("面積",area)
print("周囲長",perimeter)
print("円形度",roundness)

cnt = contours[0]

(x,y), radius=cv2.minEnclosingCircle(cnt)
center = (int(x),int(y))
radius = int(radius)
print("中心点",x,y)
print("半径",radius)

img_circle = cv2.circle(img_src,center,radius,(0,255,0),2)
cv2.imshow("dst Show contours Image",img_circle)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'daruma.png'

img_src = cv2.imread(file_src,1)#入力画像の読み込み
img_gray =cv2.cvtColor(img_src,cv2.COLOR_BGR2GRAY)


m = cv2.moments(img_gray)
area = m['m00']
x_g = m['m10'] / m['m00']
y_g = m['m01'] / m['m00']
ang = 0.5 * math.atan2(2.0 * m['mull'], m['mu20'] - m['mu02'])

#表示
print("面積",area)
print("周囲長",x_g)
print("円形度",y_g)
print("円形度2",ang)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'labeling_sample.png'

img_src = cv2.imread(file_src,1)#入力画像の読み込み
img_gray =cv2.cvtColor(img_src,cv2.COLOR_BGR2GRAY)

nlabel, img_lab = cv2.connectedComponents


L_area = []

for num in range(1,nlabel):
    img = cv2.compare(img_lab,num,cv2.CMP_EQ)

    contours = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    area = cv2.contourArea(contours[0])

    L_area.append(int(area))
    print(L_area)
    cv2.imshow("dst Show labeling Image",img)
    cv2.waitKey(0)

label_max = max(L_area)
label_min = min(L_area)
print(label_max)
print(label_min)

L_max_index=L_area.index(max(L_area))+1
L_min_index=L_area.index(min(L_area))+1
print(label_max)
print(label_min)

img_max = cv2.compare(img_lab,L_max_index,cv2.CMP_EQ)
cv2.imshow("Show labeling MAX Image",img_max)
img_min = cv2.compare(img_lab,L_min_index,cv2.CMP_EQ)
cv2.imshow("Show labeling MIN Image",img_min)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
