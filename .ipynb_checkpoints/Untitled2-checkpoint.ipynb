```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'free.jpg'
img_src = cv2.imread(file_src,1)#入力画像の読み込み
#ヒストグラム表示用,100x256ピクセル,0で初期化
img_hst = np.zeros([100,256]).astype('uint8')
rows, cols = img_hst.shape[:2]

#度数分布を求める
hdims=[256]
hranges=[0,256]
hist=cv2.calcHist([img_src],[0],None,hdims,hranges)

#度数の最大値を取得
min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(hist)
max_val=1100
for i in range(0,256):
    v=hist[i]
    print(v)
    cv2.line(img_hst,(i,rows),(i,rows-rows*(v/max_val)),(255,255,255))

cv2.imshow("src Show Input Image", img_src)#入力画像表示
#cv2.imshow("src Show Input Image", img_hst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'hana.jpg'
file_dst = 'dst_jpg'
img_src = cv2.imread(file_src,1)#入力画像の読み込み
img_gray = cv2.cvtColor(img_src,cv2.COLOR_BGR2GRAY)
#ヒストグラム表示用,100x256ピクセル,0で初期化
img_hst = np.zeros([100,256]).astype('uint8')
rows, cols = img_hst.shape[:2]
img_dst = cv2.equalizeHist(img_gray)

#度数分布を求める
hdims=[256]
hranges=[0,256]
hist=cv2.calcHist([img_dst],[0],None,hdims,hranges)
#hist=cv2.calcHist([img_gray],[0],None,hdims,hranges)

#度数の最大値を取得
min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(hist)
max_val=1100
for i in range(0,256):
    v=hist[i]
    print(v)
    cv2.line(img_hst,(i,rows),(i,rows-rows*(v/max_val)),(255,255,255))

#cv2.imshow("src Show Input Image", img_src)#入力画像表示
#cv2.imshow("src Show Input Image", img_gray)
cv2.imshow("src Show Input Image", img_hst)
#cv2.imshow("src Show Input Image", img_dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#-*- coding: utf-8 -*-
import cv2
import math
import numpy as np

file_src = 'hana.jpg'
file_dst = 'dst_jpg'
img_src = cv2.imread(file_src,1)#入力画像の読み込み

#処理記述　ガンマ変換　gamma=2.0
#ルックアップテーブル作成
gamma = 0.5
Y = np.ones((256,1), dtype = 'uint8')*0
for i in range(256):
    Y[i][0] = 255 * pow(float(i) / 255, 1.0/gamma)

#ルックアップテーブル変換
img_dst = cv2.LUT(img_src,Y)

#表示
cv2.imshow("src Show Input Image", img_src)#入力画像表示
#cv2.imshow("src Show Input Image", img_dst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
